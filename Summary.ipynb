## <span style="text-align: right; direction: rtl; float: right; clear: both;">שובו של קשטן</span>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
    בחנות של אדון קשטן ישנו מילון המכיל את כל המוצרים בחנות, ואת המלאי של כל מוצר.<br>
    רשמו פונקציה שמקבלת את המוצרים שהתקבלו במשלוח האחרון, ומעדכנת את המלאי של אדון קשטן בהתאם לכך.
</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
לדוגמה:
</p>

<ul style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>עבור הקריאה <code dir="ltr" style="direction: ltr; text-align: left;">add_inventory({'cheese': 2, 'milk': 1}, cheese=3, chocolate=5)</code><br>
        יוחזר המלאי: <samp dir="ltr" style="direction: ltr; text-align: left;">{'cheese': 5, 'chocolate': 5, 'milk': 1}</samp>
    </li>
    <li>עבור הקריאה <code dir="ltr" style="direction: ltr; text-align: left;">add_inventory({'refrigerator': 7, 'goat': 1}, honey=2)</code><br>
    יוחזר המלאי: <samp dir="ltr" style="direction: ltr; text-align: left;">{'honey': 2, 'refrigerator': 7, 'goat': 1}</samp>
    </li>
</ul>

```python
def add_inventory(current_inventory, **products):
    for product_key in products:
        if product_key in current_inventory:
            temp = int(current_inventory[product_key]) + int(products[product_key])
            current_inventory[product_key] = temp
        else:
            current_inventory[product_key] = products[product_key]
    return current_inventory

```

```python
current_inventory = add_inventory({'cheese': 2, 'milk': 1}, cheese=3, chocolate=5)
print(f"current_inventory: {current_inventory}")
```

```python
current_inventory = add_inventory({'refrigerator': 7, 'goat': 1}, honey=2)
print(f"current_inventory: {current_inventory}")
```

```python
from os import listdir, rename
import codecs
from bs4 import BeautifulSoup

def reorganize__chapters(path):
    list_of_files = listdir(path)
    for file_name in list_of_files:
        print(file_name)
        # openning the file
        file_data = codecs.open(path + '\\' + file_name, 'r', encoding="utf-8")
#         file_data = open(path + '\\' + file_name, 'r')
        
        #finding the title
        soup = BeautifulSoup(file_data, 'html.parser')  
        title = str(soup.find("title"))[52:]
        title = title[:-8]
        title = title.replace(':','-')
        print(title)
        
        #closing the file
        file_data.close()
        
        #renaming the file
        current_name = path + '\\' + file_name
        print(f"current_name: {current_name}")
        new_name = path + '\\' + title + '.html'
        print(f"new_name: {new_name}")
        rename(current_name, new_name)
```

<blockquote style="border-left:none !important; position: relative; padding-left: 0.5em; padding: 1.5em; border-right: 3px solid #a93226; background-color: #fbe7e6; direction: rtl; text-align: right; clear: both;">
    מה הסיכוי שבכיתה שבה 23 תלמידים (חחחחח) יציינו שני תלמידים יום הולדת באותו תאריך?<br>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
הנחות המוצא שלנו בפתירת הבעיה הן כדלהלן:
</p>

<ul style="text-align: right; direction: rtl; float: right; clear: both;">
    <li>הילודה מפוזרת באופן שווה לאורך השנה (<a href="https://en.wikipedia.org/wiki/Birthday#Frequency">היא לא</a>, אבל נדמיין שכן).</li>
    <li>כשאנחנו אומרים "תאריך" אנחנו מתכוונים לצירוף של החודש והיום, ללא התייחסות לשנה.<br>
        מי שנולד ב־1 במאי 1992 הוא בעל תאריך יום הולדת זהה למי שנולד ב־1 במאי 1995.</li>
</ul>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
    הפתרון של ה"פרדוקס" מפתיע את רוב האנשים ששומעים עליו לראשונה.
</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
    כתבו תוכנה שמגרילה תאריכי יום הולדת עבור 10,000 כיתות, שבכל אחת מהן לומדים 23 תלמידים.<br>
    מהו אחוז הכיתות שבהן ציינו 2 ימי הולדת באותו תאריך? השוו עם הפתרון המוצג בוויקיפדיה.<br>
    דאגו שהפתרון שלכם יתחשב בכך שקיימות שנים מעוברות.<br>
</p>

<p style="text-align: right; direction: rtl; float: right; clear: both;">
    השתמשו בתוכנה שכתבתם כדי לחשב את הסיכויים עבור כיתות שבהן תלמיד אחד, ועד כיתות ישראליות ממוצעות שבהן 366 תלמידים.
</p>

```python
import random

def random_student_birthdays(num_of_students_in_class):
    i = 1
    students_birthdays = []
    for i in range(num_of_students_in_class):
        students_birthdays.append(int(random.randint(1,366)))
    return students_birthdays

def check_if_there_is_duplicates(students_birthdays):
    for birthday in students_birthdays:
        if students_birthdays.count(birthday) > 1:
            return True
    return False
    
def creates_database(num_of_students_in_class):
    database = []
    for i in range(1,10000):
        database.append(check_if_there_is_duplicates(random_student_birthdays(num_of_students_in_class)))
    return database
    
def calculate_probability(database):
    return database.count(True) / len(database)

def check_paradox_birthday(num_of_students_in_class):
    database = creates_database(num_of_students_in_class)
    probability = calculate_probability(database)
    return probability
    
def check_paradox_birthday_of_variety_of_classes():
    for num_of_students_in_class in range(1,366):
        probability = check_paradox_birthday(num_of_students_in_class)
        print(f"In a class of {num_of_students_in_class} students the probability is {probability}")
```
